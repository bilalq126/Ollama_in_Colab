```python
from ollama import chat

resp = chat(model='mistral', messages=[{'role': 'user', 'content': 'Write a 2-line poem about Colab.'}])
print(resp['message']['content'])
```

```python
# streaming
from ollama import chat
for chunk in chat(model='mistral', messages=[{'role':'user','content':'Tell a short joke.'}], stream=True):
    print(chunk['message']['content'], end='', flush=True)
```

```python
from pyngrok import ngrok
# replace with your ngrok authtoken if required:
# ngrok.set_auth_token("YOUR_NGROK_AUTHTOKEN")
t = ngrok.connect(11434, "http")
print("Public URL:", t.public_url)
# Use the public URL as host when creating the Python client:
from ollama import Client
client = Client(host=t.public_url)   # host must include http(s) prefix
resp = client.chat(model='mistral', messages=[{'role':'user','content':'Hello via ngrok'}])
print(resp.message.content)
```
